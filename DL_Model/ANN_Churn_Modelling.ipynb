```python
df = pd.read_csv('Churn_Modelling.csv')
df
```

```python
df = pd.get_dummies(df, columns=['Geography','Gender'], drop_first=True)
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'], inplace=True)
df
```

```python
X = df.drop(columns=['Exited'])
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scalled = scaler.fit_transform(X_train)
X_test_scalled = scaler.transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Dropout, BatchNormalization
```

```python
model = Sequential()

model.add(Dense(7, activation='relu', input_dim = 11))
#model.add(Dropout(0.6))
model.add(Dense(5, activation='relu'))
model.add(Dense(7, activation='relu'))
#model.add(BatchNormalization())
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train_scalled, y_train, epochs=35, validation_split=0.2)
```

```python
X_train_scalled
```

```python
model.layers[0].get_weights()
```

```python
y_prob = model.predict(X_test_scalled)
y_prob
```

```python
y_pred = np.where(y_prob>0.5,1,0)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
plt.plot(history.history['loss'], label='loss', color='r')
plt.plot(history.history['val_loss'], label='val_loss', color='g')
plt.legend()
```

```python
plt.plot(history.history['accuracy'], label='accuracy', color='r')
plt.plot(history.history['val_accuracy'], label='val_accuracy', color='g')
plt.legend()
```
